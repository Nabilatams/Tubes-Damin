```python
# Statistik Kategorikal
print("\nStatistik Deskriptif Kolom Kategorikal:")
display(df.describe(include='object'))

categorical_columns = [
    'First_Name', 'Last_Name', 'Gender', 'Department', 'Grade',
    'Extracurricular_Activities', 'Internet_Access_at_Home',
    'Parent_Education_Level', 'Family_Income_Level'
]

for col in categorical_columns:
    print(f"\n{col}:")
    display(df[col].value_counts().reset_index().rename(columns={'index': col, col: 'Jumlah'}))
```
