```python
## üìä Dashboard Overview

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "data/Students Performance Dataset.csv"
df = pd.read_csv(file_path)

# Tampilkan 5 data teratas
print("\n‚ñ∂Ô∏è 5 Baris Pertama:")
display(df.head())

# Jumlah data
print("\n‚ñ∂Ô∏è Jumlah Data:")
print(f"Baris: {df.shape[0]}, Kolom: {df.shape[1]}")

# Distribusi nilai kelulusan jika sudah ada kolom 'Class'
if "Class" in df.columns:
    print("\n‚ñ∂Ô∏è Distribusi Kelas:")
    display(df["Class"].value_counts())
    df["Class"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, colors=["red", "green"])
    plt.title("Distribusi Class (Pass/Fail)")
    plt.ylabel("")
    plt.show()

# Statistik umum
print("\n‚ñ∂Ô∏è Statistik Umum:")
display(df.describe())
```
